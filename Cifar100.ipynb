```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D, AveragePooling2D
import pickle
import numpy
```

```python
with open('D:/cifar-100-python/train', 'rb') as fo:
    train = pickle.load(fo, encoding='bytes')
with open('D:/cifar-100-python/test', 'rb') as fo:
    test = pickle.load(fo, encoding='bytes')
test.keys()
```

```python
train_features = train[b'data'].reshape(train[b'data'].shape[0],
                                           32, 32, 3)
train_coarse_labels = numpy.array(train[b'coarse_labels'],
                                      dtype=numpy.uint8)
train_fine_labels = numpy.array(train[b'fine_labels'],
                                    dtype=numpy.uint8)
train_coarse_labels
```

```python
test_features = test[b'data'].reshape(test[b'data'].shape[0],
                                         32, 32, 3)
test_coarse_labels = numpy.array(test[b'coarse_labels'], dtype=numpy.uint8)
test_fine_labels = numpy.array(test[b'fine_labels'], dtype=numpy.uint8)
test_fine_labels
```

```python
x_train = train_features/255.0
x_test = test_features/255.0
```

```python
from keras.utils import np_utils
y_train = np_utils.to_categorical(train_fine_labels.reshape(-1,1))
y_test = np_utils.to_categorical(test_fine_labels.reshape(-1,1))
num_classes = y_test.shape[1]
```

```python
from keras.constraints import maxnorm
from keras.layers.normalization import BatchNormalization
model = Sequential()
model.add(Conv2D(16, (3, 3), input_shape=x_train.shape[1:], activation='relu', padding='same'))
model.add(Conv2D(63, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(Dropout(0.2))
model.add(Conv2D(63, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(63, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(Dropout(0.2))
model.add(Conv2D(63, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(63, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(Dropout(0.2))
model.add(Conv2D(63, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(63, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(0.2))
model.add(Conv2D(126, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(Dropout(0.2))
model.add(Conv2D(126, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(126, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(Dropout(0.2))
model.add(Conv2D(126, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(126, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(Dropout(0.2))
model.add(Conv2D(126, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(126, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(0.2))
model.add(Conv2D(252, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(Dropout(0.2))
model.add(Conv2D(252, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(252, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(Dropout(0.2))
model.add(Conv2D(252, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(252, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(Dropout(0.2))
model.add(Conv2D(252, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(252, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(BatchNormalization())
#model.add(Activation('tanh'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(1308, activation='relu', kernel_constraint=maxnorm(3)))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))
model.summary()
```

```python
'''from keras.optimizers import SGD
sgd = SGD(lr=0.0001, momentum=0.9, decay=0.0001/25, nesterov=False)'''

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x_train, y_train,validation_split=0.25, epochs=200, batch_size=32)
#model.summary()
```

```python
scores = model.evaluate(x_test,y_test,verbose=1)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
#plt.show()
```
